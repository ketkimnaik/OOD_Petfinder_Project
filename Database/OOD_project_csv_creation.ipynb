```python
import pandas as pd
import requests, bs4
import numpy as np
```

```python
df = pd.read_csv('happytail_raw_data.csv')
arr = df["url"].to_numpy()
arr.size
```

```python
l = []
for url in arr:
    headers = {"User-Agent":"Mozilla/5.0"}
    response = requests.get(url, headers=headers)
    soup = bs4.BeautifulSoup(response.text, 'lxml')
    images = soup.findAll('img')
    temp=""
    for image in images:
        temp = image['src']
    l.append(temp)
```

```python
print(len(l))
```

```python
df['pic_new'] = l
print(df)
```

```python
df.to_csv('new_happy_tail_raw_data')
```
